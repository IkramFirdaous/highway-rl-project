```python
import json
import matplotlib.pyplot as plt
```

```python
with open('results/metrics/dqn_seed44.json', 'r') as f:
    reward_json = json.load(f)

print(reward_json)
```

```python
def plot_training_curves(reward,seed=44):
    plt.figure()
    plt.plot(reward, alpha=0.7, label=f"seed={seed}")
    plt.xlabel("Episode")
    plt.ylabel("Reward")
    plt.title("DQN training curves")
    plt.legend()
    plt.tight_layout()
    plt.savefig("results/plots/dqn_training_curves.png")
    plt.close()


plot_training_curves(reward_json['rewards'])
```
